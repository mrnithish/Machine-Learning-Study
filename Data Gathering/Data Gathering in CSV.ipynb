## Converters
It is used to pass the function

```python
def rename(name):
    if name=="Royal Challengers Bangalore":
        return 'RCB'
    else:
        return
```

```python
pd.read_csv("matches.csv",converters={'team1':rename})
```

## na_values parameter

```python
pd.read_csv("matches.csv",na_values=['Royal Challengers Bangalore'],nrows=10)
```
