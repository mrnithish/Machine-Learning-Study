# Univariate Analysis

```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("train.csv")
df.head()
```

## Categorical Data

### CountPlot

It is used find the frequency

```python
dfs = pd.DataFrame(df['Survived'])
dfs
sns.countplot(x='Survived',data=dfs)
```

```python
df['Pclass'].value_counts().plot(kind='bar')
```

```python
df['Survived'].value_counts().plot(kind='bar')
```

## Pie Chart

```python
df['Survived'].value_counts().plot(kind='pie',autopct='%.2f')
```

## Numerical Data

### Histogram

Find range of time
Distribution of data in numerial  range

```python
import matplotlib.pyplot as plt
plt.hist(df['Age'])
```

### Distplot

```python
sns.distplot(df['Age'])
```

### Boxplot

find outliers and noisy data

```python
sns.boxplot(df['Fare'])
```

```python
df['Age'].min()
```

```python
df['Age'].skew() # Find whether it is  symmentrical or not
```

```python
df['Fare'].mean()
```
